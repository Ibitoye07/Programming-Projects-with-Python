## Question 1

```python
# storing scores in a list
james_scores= [78,90,67,52,84]
```

```python
# determining his highest and lowest scores
highest_score= max(james_scores)
lowest_score= min(james_scores)
highest_score

print('The highest score is:', highest_score)
print('The lowest score is:', lowest_score)
```

```python
# Finding the average of his scores
Average_score= sum(james_scores)/len(james_scores)
print('Average score is:', Average_score )
```

```python
print (f"James's best score is {highest_score}, and his lowest score is {lowest_score}. His average score is {Average_score}")
```

## Question 2

```python
# store amount in a list
Amount= [1200,1000,450,780,870,1100,1500]
```

```python
# determining the most and least expensive amount
most_expensive_amount= max(Amount)
least_expensive_amount= min(Amount)

print('Most expensive amount is:', most_expensive_amount)
print('least expensive amount is:', least_expensive_amount)
```

```python
# sorting amount list in descending order
ordered_amount= sorted(Amount, reverse=True)
top_3= ordered_amount[:3]
```
