## Student Grading System

```python
# Assigning grades using conditional statements and function
def assign_grade (score):
    if score >=90:  # score is greater than or equal to 90
        return 'A'  # grade to display
    
    elif score >=80: # score is greater than or equal to 80
        return 'B'   
    
    elif score >=70: # score is greater than or equal to 70
        return 'C'
    
    elif score >=60: # score is greater than or equal to 60
        return 'D'
    
    elif score >=0: # score is greater than or equal to 0
        return 'F'  
    
    else:           # initiate when it does not fulfil the above conditions i.e score is below 0
        return 'Score is out of scope'

# User is required to input score
user_score = int(input('Enter your score:'))

# get the grade
grade= assign_grade (user_score)

# display grade performance
print(f'Your grade performance is: {grade}')
```

## Simple Number Game

```python
# import library
import random
import emoji

# generate a random secret number between 1 and 10
secret_number= random.randint(1,10)

# set initial guess to 0 before attempting a guess
guess= 0
# set counter to be 0
counter=0

# loop until the user guesses the secret number
while guess!=secret_number:
    guess= int(input('\nEnter your guess:'))
   
    # counter increment
    counter+=1 
    
    # Check to see if user has exceeded the number of chances given, if so Terminate the loop
    if counter >3:
        print('No result displayed, you have exceeded all three chances')
        break  # Exit the loop
    
    # Check to see if you guessed the secret number correctly
    if guess == secret_number:
        print ('CONGRATULATIONS!!! You win', emoji.emojize(':thumbs_up::champagne:', language='alias'))
    else:
        print ('OOps, you lose..TRY AGAIN', emoji.emojize(':smile:', language='alias'))
```

## Multiplication Table Generator

```python
# initialize function
def generate_table():

    # Ask the user to input a number
    num= int(input('Enter a number:'))
    
    # display heading
    print(f'\nMultiplication Table of {num}:\n')
    
    # number of times to loop
    for i in range(1,13):
        print(f'{num} x {i} = {num*i}') # display the results 
 
# call the function
generate_table()
```
